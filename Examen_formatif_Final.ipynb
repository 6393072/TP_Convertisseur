### A4) Mini‑analyse YAML (bonus +2 pts)  
Considérez cet extrait :

```yaml
name: Python Tests
on: [push, pull_request]

jobs:
  test:
    runs-on: ubuntu-latest
    steps:
      - name: Cloner le dépôt
        uses: actions/checkout@v3
      - name: Installer Python
        uses: actions/setup-python@v4
        with:
          python-version: '3.10'
      - name: Installer les dépendances
        run: pip install -r requirements.txt
      - name: Lancer les tests
        run: pytest tests/
```

## Exercice 1 (/3 pts)

Soit le bloc de code suivant :

```python
# Code de la fonction à tester : pour calculer la somme des diviseurs inférieurs à n
def somme_quoi_inferieurs(n):
    somme = 0
    for i in range(1,n/2,1):
        if n % i == 0:
            somme += i
```

### Travail demandé :
Complétez les **tests unitaires** pour cette fonction.

- Tester un nombre premier  
- Tester un nombre parfait (ex: 6)  
- Tester un cas limite (n = 1)  
- Tester un type invalide (si pertinent)  

Utilisez `pytest` ou `unittest`.


````python
# Code de la fonction à tester : pour calculer la somme des diviseurs inférieurs à n
def somme_quoi_inferieurs(n):
    somme = 0
    for i in range(1,n//2,1):
        if n % i == 0:
            somme += i
    return somme

import unittest
class Test_somme(unittest.TestCase):
    def test_nombre_premier(self):
        self.assertEqual(somme_quoi_inferieurs(5),1)
    def test_nombre_parfait(self):
        self.assertEqual(somme_quoi_inferieurs(6),3)
    def test_cas_limite(self):
        self.assertEqual(somme_quoi_inferieurs(1),0)
    def test_nombre_negatif(self):
        self.assertEqual(somme_quoi_inferieurs(-12),0)

unittest.TextTestRunner().run(unittest.TestLoader().loadTestsFromTestCase(Test_somme))
````

## Exercice 2 (/5 pts)

Soit le bloc de code suivant :

```python
# Créez une fonction simple à tester utiliser la classe TestCase.
def addition(a,b):
    return a+b-1

# Test unitaire sans la classe
def test_addition():
    result = addition(2,3)
```

### Travail demandé :
Ajoutez les instructions de test unitaire au code en utilisant la classe `TestCase`.

- Créez une classe héritant de `unittest.TestCase`
- Écrivez au moins 3 tests
- Montrez qu’un test échoue à cause de l’erreur logique dans `addition`
- Proposez la correction de la fonction


````python
# Créez une fonction simple à tester utiliser la classe TestCase.
def addition(a,b):
    return a+b-1
# Fonction corrigé
def vrai_addition(a,b):
    return a+b
    
# Classe héritant unittest
import unittest
class TestAddition(unittest.TestCase):
    # Test qu'échoue à cause de l'érreur logique dans la fonction
    def test_addition(self):
        self.assertEqual(addition(2,3),5)
    # Tests du fonction corrigé
    def test_vrai_addition(self):
        self.assertEqual(vrai_addition(2,3),5)
    def test_vrai_addition_zero(self):
        self.assertEqual(vrai_addition(0,0),0)
    def test_vrai_addition_negatif(self):
        self.assertEqual(vrai_addition(-52,-13),-65)

unittest.TextTestRunner().run(unittest.TestLoader().loadTestsFromTestCase(TestAddition))
````
